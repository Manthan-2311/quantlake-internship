# Day 6 - Data Visualization with Matplotlib & Seaborn
Learn how to visualize and communicate data using plots.

## Load Data and Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('superstore.csv', encoding='ISO-8859-1')
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("-", "_")
df.head()
```

## Task 1: Matplotlib Basics

```python
# Line Plot
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.title("Simple Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.savefig("fig1_lineplot.png", dpi=300)
plt.show()

# Scatter Plot
plt.scatter([1, 2, 3], [3, 5, 1])
plt.title("Simple Scatter Plot")
plt.savefig("fig2_scatter.png", dpi=300)
plt.show()
```

## Task 2: Plot Customization

```python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], label='Sales')
plt.title("Customized Line Plot")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.grid(True)
plt.legend()
plt.savefig("fig3_customized_plot.png", dpi=300)
plt.show()
```

## Task 3: Seaborn Visualizations

```python
# Boxplot
sns.boxplot(data=df, x='category', y='profit')
plt.title("Profit by Category")
plt.savefig("fig4_boxplot.png", dpi=300)
plt.show()

# Heatmap
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.savefig("fig5_heatmap.png", dpi=300)
plt.show()

# Histogram
sns.histplot(df['sales'], bins=30, kde=True)
plt.title("Sales Distribution")
plt.savefig("fig6_histogram.png", dpi=300)
plt.show()
```

## Task 4: Mini Visualization Project

```python
# Histogram of Sales
sns.histplot(df['sales'], bins=40)
plt.title("Sales Histogram")
plt.savefig("fig7_sales_histogram.png", dpi=300)
plt.show()

# Boxplot of Profit by Category
sns.boxplot(x='category', y='profit', data=df)
plt.title("Profit by Category")
plt.savefig("fig8_profit_boxplot.png", dpi=300)
plt.show()
```

## Task 5: Final Chart - Polished Figure

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='region', y='sales', estimator=sum, ci=None)
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("final_chart.png", dpi=300)
plt.show()
```
